##### Python Collections (Arrays)
There are four collection data types in the Python programming language:

List is a collection which is ordered and changeable. Allows duplicate members.

Tuple is a collection which is ordered and unchangeable. Allows duplicate members.

Set is a collection which is unordered and unindexed. No duplicate members.

Dictionary is a collection which is ordered* and changeable. No duplicate members.

```python
#import required libraries
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
from math import sqrt
```

```python
#create x/y co - ordinates (data points)
X = [1,3,4,5,1,4]
Y = [3,3,3,3,2,2]

#centroid 1 co-ordinates
cx1 = 1
cy1 = 1
#centorid 2 co-ordinates
cx2 = 2
cy2 = 1

#initiliaze empty lists
list1 = []
list2 = []
list3 = []
list4 = []
list5 = []
list6 = []

#eulcidean distance calculations
for xi in X:
    while len(X) <= len(X):
        list1.append((cx1 - xi) ** 2)
        break
for yi in Y:
    while len(Y) <= len(Y):
        list2.append((cy1 - yi) ** 2)
        break
        
##eulcidean distance calculations continued

#for loop to loop over list
for xii in X:
    #while loop : expression
    while len(X) <= len(X):
        #appending(adding calculations to list)
        list3.append((cx2 - xii) ** 2)
        #break out of the while loop
        break
for yii in Y:
    while len(Y) <= len(Y):
        list4.append((cy2 - yii) ** 2)
        break
        
#for loop zipping together two lists   
for x,y in zip (list1,list2):
    while len(list1) <= len(list2):
        #square root calculations
        phase1 = sqrt(x + y)
        list5.append(round(phase1,2))
        break
        
for a,b in zip (list3,list4):
    while len(list3) <= len(list4):
        phase2 = sqrt(a + b)
        list6.append(round(phase2,2))
        break

print(X,Y)
print('Euclidean Distance Calculations:',list5)
#print(type(list5))
print('Eulclidean Distance Calculations:',list6)
#print(type(list6))

#Transfering the above lists into a Dictionary
Dictionary = {'X':X, 'Y':Y, 'Calculations Centroid1':list5, 'Calculations Centroid2':list6}
#Dictionary

#Create a DataFrame from our above Dictionary
df = pd.DataFrame(Dictionary)
#Create copy of original dataframe
original_df = df.copy(deep=True)

#Labelling the clusters for interpretability
Clusters = []

#Assign Labels for Centroid/Clusters
label1 = 'Centroid1'
label2 = 'Centroid2'

#initiate for loop to assign labels
#NOTE: These labels will represent which Cluster X & Y co-ordinates belong too !
for i in df['Calculations Centroid1']:
    for f in df['Calculations Centroid1']:
        if i > f:
            Clusters.append(label1)
        if i <= f:
            Clusters.append(label2)
        break
print('Cluster Labels :', Clusters)

#add column too dataframe
df['ClusterLabels'] = Clusters
df
```
